```python
#მონაცემთა ბაზის წაკიტხვა
import pandas as pd
df = pd.read_csv("Earthquakes_database.csv")
df
```

# დავალება 1. მონაცემთა წმენდის ფუნქციონალი გამოყენებულია split, drop, fillna

```python
#Time სვეტი გადანაწილებულია სამ ახალ სვეტად: Hour, Minute, Second
df[['Hour', 'Minute', 'Second']] = df['Time'].str.split(':', expand=True)
df
```

```python
#Source სვეტის წაშლა
df = df.drop(columns = "Source")
df
```

```python
# Horizontal Distance სვეტში NaN ცარიელი მნიშვნელობების შეცვლა 0-ით
df["Horizontal Distance"].fillna(0, inplace=True)
df
```

# დავალება 2. რიცხვითი ტიპის ველის მონაცემების დაყოფა კატეგორიებად cut() ფუნქციის გამოყენებით

```python
# Magnitude სვეტისთვის ბინების განსაზღვრა
magnitude_bins = [3.9, 4.9, 5.9, 6.9, 7.9, 8.0, 10] 
# Magnitude სვეტისთვის კატეგორიების განსაზღვრა
magnitude_labels = ['Minor', 'Light', 'Moderate', 'Strong', 'Major', 'Great'] 
#შესაბამისი კატეგორიების Magnitude Category სვეტში დამატება
df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=magnitude_bins, labels=magnitude_labels, right=False) 

df
```

1. სვეტოვანი დიაგრამა bar()

```python

import matplotlib.pyplot as plt

# პირველი 10 მნიშვნელობის წამოღება Time და Magnitude სვეტებიდან
date = df['Time'].head(20)
magnitude = df["Magnitude"].head(20)

fig, ax = plt.subplots(figsize=(13, 7))

# სვეტოვანი დიაგრამის აგება
ax.bar(date, magnitude, color='blue', width=0.3)

# ლეიბლების მითითება
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Source', fontsize=14)
plt.xticks(rotation='vertical')


# დიაგრამის 
plt.show()
```

2. წრიული დიაგრამა pie()

```python
import matplotlib.pyplot as plt

# პირველი 10 მნიშვნელობის წამოღება Magnitude სვეტიდან
magnitude_sum = df["Magnitude"].tail(10).sum()

fig, ax = plt.subplots(figsize=(7, 7))

# წრიული დიაგრამის აგება
ax.pie(df["Magnitude"].tail(10), labels=df['Source'].tail(10), autopct='%1.1f%%', startangle=90)

# სიაგრამის ჩვენება
plt.show()
```

3. ხაზოვანი დიაგრამა plot()

```python
import matplotlib.pyplot as plt

#პირველი 20 მნიშვნელობის წამოღება Date და Magnitude სვეტებიდან
date = df['Date'].tail(40)
magnitude = df['Magnitude'].tail(40)

#ხაზოვანი დიაგრამის შექმნა
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(date, magnitude, marker='o', linestyle='-', color='green')

#სათაურის და ლეიბლების განსაზღვრა
ax.set_title('Magnitude Line Chart')
ax.set_xlabel('Date')
ax.set_ylabel('Magnitude')
plt.xticks(rotation='vertical')

#დიაგრამის ჩვენება
plt.show()
```

4. ჰისტოგრამა hist()

```python
import matplotlib.pyplot as plt

#პირველი 20 მნიშვნელობის წამოღება Magnitude სვეტიდან
magnitude_type_data = df["Magnitude Category"]

# Create a histogram
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(magnitude_type_data, bins=20, color='skyblue', edgecolor='black')

# Set labels and title
ax.set_xlabel('Magnitude', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
ax.set_title('Histogram of Earthquake Magnitudes', fontsize=16)

# Rotate x-axis labels vertically
plt.xticks(rotation='vertical')

# Display the plot
plt.show()
```

# დავალება 5. თარიღებთან მუშაობა

```python
import pandas as pd

#Date სვეტის მნიშვნელობები გადადის datetime ფორმატში
df.loc[:, 'Date'] = pd.to_datetime(df['Date'], errors='coerce')

#დაფორმატებული Date სვეტიდან იღებს Nan მნიშვნელობებს
df = df.dropna(subset=['Date'])

#ვამატებთ ახალ სვეტებს day, month და year
df.loc[:, 'Day'] = df['Date'].dt.day
df.loc[:, 'Month'] = df['Date'].dt.month
df.loc[:, 'Year'] = df['Date'].dt.year

df
```
